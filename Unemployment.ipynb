# loading Libraries and reading data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("unemployment analysis.csv")
```

```python
#checking the first 5 rows
df.head(5)
```

```python
#slicing out Nigeria Only!
df[df['Country Name']=='Nigeria']
```

```python
#tranposing the rows to column to get the year and values by side 
df = df[df['Country Name']=='Nigeria']
df = np.transpose(df)
```

```python
#slicing the data from 2015 to 2021
df = df.iloc[26:]

#reseting the index to default
data = df.reset_index()

#renaming the Colummn name from 'index' to 'Year' and '152' to 'Value' 
dictRename = {
    'index': 'Year',
    152: 'Value'
}
data.rename(columns = dictRename,inplace=True)
```

```python
#checking the datatype fro each row
data.dtypes

#Changing the datatypes for each row
data['Year'] = data['Year'].astype('object')
data['Value'] = data['Value'].astype('float')
```

# ploting chart

```python
plt.plot(data["Year"],data["Value"], color = 'red', marker = '.')
plt.yticks([4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10])
sns.set_style('dark')
plt.grid()
plt.title('Unemployment in Nigeria from 2015 to 2021')
plt.xlabel('Year', fontsize= 12)
plt.ylabel('values in %', fontsize = 12)
plt.savefig('NigUN.jpeg')
```
